```python
#fig1, ax1 = plt.subplots(5,figsize=(10,50))
x1=(-3,3)
y1=(0,0)
x2=(0,0)
y2=(-3,3)
for yr in range(1981,2019,1):
    #print(yr)
    dStart=datetime.datetime(yr-1, 12, 1)
    #print(dStart)
    dEnd=datetime.datetime(yr, 1, 31)
    dEnd2=datetime.datetime(yr, 1, 31,9)
    #print(dEnd)
    obs1 = obs.sel(time=slice(dStart,dEnd))
    ao1  = ao.sel(time=slice(dStart,dEnd2))
    eu1  = eu.sel(time=slice(dStart,dEnd2))
    cList=[]
    for f in obs1 :
        if(f == 0):
            cList.append('Green')
        elif (f < 0.48):
            cList.append('Blue')
        elif (f < 0.6):
            cList.append('red')
        elif (f >= 0.6):
            cList.append('black')
    scatter=plt.scatter(ao1, eu1, s=25, c=cList )
    plt.xlim(-3, 3)
    plt.ylim(-3, 3)
    plt.plot(x1,y1, linestyle = ":", lw=1,color = 'black')
    plt.plot(x2,y2, linestyle = ":", lw=1,color = 'black')
    # produce a legend with the unique colors from the scatter
    plt.legend()
    #plt.legend
    #plt.add_artist(legend1)
    #plt.plot([-3,0],[3,0], linestyle = ":", lw=1)
    #plt.xlim([0, 3])
    #plt.ylim([0, 3])
    plt.title(yr)
    plt.pause(0.05)
    plt.show()    
    #print(obs1.time)
    #print(ao1.time)
#plt.show()    
```
