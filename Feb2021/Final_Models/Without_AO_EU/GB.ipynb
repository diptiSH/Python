```python
import numpy as np
import pandas as pd
import xarray as xr
import math as mt
from joblib import dump, load
import datetime
from sklearn.model_selection import train_test_split
from random import seed
from random import randint
import statistics
from scipy.stats import stats
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import  RandomizedSearchCV
from time import time
from scipy.stats import uniform
from sklearn.utils.fixes import loguniform
import matplotlib.pyplot as plt
from scipy.stats import randint
```

# Load Input Data

```python
# Load Inputs
# Load X and y values
X = load('X.joblib') 
y= load('y.joblib')
```

# Randomaly  Spit train test (20%) 

```python
X_train,X_test,y_train,y_test= train_test_split(X ,y,test_size=0.20, random_state=0)
```

# Hyper Parameter tuning on train data 10fold CV and RandomizedSearchCV

```python
gbReg = GradientBoostingRegressor()


# Utility function to report best scores
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})"
                  .format(results['mean_test_score'][candidate],
                          results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")

# specify parameters and distributions to sample from
param_dist = {'loss': ['ls','lad'],
              'max_depth': randint(1, 5),
              'n_estimators':randint(100,1000) ,
              'min_samples_leaf':randint(5,25) ,
              'learning_rate': loguniform(1e-3, 1e-1)
             }

# run random search
n_iter_search = 20
random_search = RandomizedSearchCV(gbReg, param_distributions=param_dist,scoring='r2',cv=10,n_iter=n_iter_search)
start = time()
random_search.fit(X_train, y_train)

print("randomSearchCV took %.2f seconds for %d candidate parameter settings."
      % (time() - start, len(random_search.cv_results_['params'])))
report(random_search.cv_results_)
```

# Fit model with selected hyperparameters on train data and test it on test data

```python
regLin=GradientBoostingRegressor(n_estimators=500, min_samples_leaf=10, max_depth=4,learning_rate=0.01,random_state=0).fit(X_train, y_train)
#regLin=GradientBoostingRegressor(n_estimators=250, min_samples_leaf=11, max_depth=4,learning_rate=0.05,random_state=0).fit(X_train, y_train)
acc_svr       =regLin.score(X_test, y_test) 
print("Coefficient of determination of base SVR model on test data is ,",(acc_svr))
yy_test = regLin.predict(X_test)
test_corr = np.corrcoef(yy_test,y_test)[0,1]
plt.title(' SVR corr = %f'%test_corr)
plt.plot(y_test,yy_test,'+')
```

# Predict all data

```python
yLR=regLin.predict(X)
y_predLin_ds=xr.Dataset({'yLR': (('time'), yLR)}, coords={'time': y.time})
dump(y_predLin_ds.yLR,'TSdataGB.joblib')
```

# PLot TS 

```python

plt.figure(figsize=[15,5])
y.resample(time='1Y').mean().plot(label='Obs')
y_predLin_ds.resample(time='1Y').mean().yLR.plot(label='Pred_GB')
dump(y,'TSdataObs.joblib')
plt.ylim(0, 0.6)
plt.legend()
```
