```python
import xarray as xr
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
from math import e
from joblib import dump, load
import statistics as st
from calendar import isleap
```

```python
N2OHistvals=[]
# N2O Data
N2ODataIPSL = xr.open_dataset('../../Data/CMIP6PollutionData/n2oglobal_Amon_IPSL-CM6A-LR_historical_r1i1p1f1_gr_185001-201412.nc')
N2OTS1=N2ODataIPSL.n2oglobal
N2OTs=(N2OTS1[((N2OTS1.time.dt.month>11) | (N2OTS1.time.dt.month<2)) ].sel(time=slice('1980-12-1','2014-1-31')))
x=np.arange(1981,2014)
histN2OTs=[]
for yr in x :
    InDString=np.datetime64(str(yr-1)+'-12-01T12:00:00')
    EnDString=np.datetime64(str(yr)+'-01-31T12:00:00')
   # print(EnDString)
    N2OT=N2OTs.sel(time=slice(InDString,EnDString))
   # print(rhT.time)
    histN2OTs.append(N2OT.mean())
N2OHistvals.append(histN2OTs) 

fig = plt.figure(figsize=(10,7))
plt.plot(x,N2OHistvals[0],label='IPSL',color='Red')
plt.legend()

plt.savefig('HistN2O_Anthropogenic.eps')
plt.show()
```

```python
ssps=['126','245','370','585']
fig = plt.figure(figsize=(22.5,5))

i=1
for sp in ssps:
    
    x=np.arange(1981,2014)

    plt.subplot(1, 4, i)

    #plt.plot(x,EnsBCHist,label='Historical',color='gainsboro')
    plt.plot(x,N2OHistvals[0],color='lightcoral')
    #plt.plot(x,N2OHistvals[1],color='cornflowerblue')
    x=np.arange(2016,2101)
    N2OFut=[]
# N2O Data
    N2ODataIPSL = xr.open_dataset('../../Data/CMIP6PollutionData/n2oglobal_Amon_IPSL-CM6A-LR_ssp'+sp+'_r1i1p1f1_gr_201501-210012.nc')

    N2OTS1=N2ODataIPSL.n2oglobal

    N2OTs=(N2OTS1[((N2OTS1.time.dt.month>11) | (N2OTS1.time.dt.month<2)) ].sel(time=slice('2015-12-1','2100-1-31')))
    #x=np.arange(1981,2014)
    histN2OTs=[]
    for yr in x :
        InDString=np.datetime64(str(yr-1)+'-12-01T12:00:00')
        EnDString=np.datetime64(str(yr)+'-01-31T12:00:00')
   # print(EnDString)
        N2OT=N2OTs.sel(time=slice(InDString,EnDString))
   # print(rhT.time)
        histN2OTs.append(N2OT.mean())
    N2OFut.append(histN2OTs) 

    
    plt.plot(x,N2OFut[0],color='lightcoral')
    jointList=N2OHistvals[0]+N2OFut[0]
    S = pd.Series (jointList)
    xTotal=np.arange(1982,2100)
    rENS=S.rolling(10, center=True, min_periods=1).mean()
    plt.plot(xTotal,rENS,label='IPSL-CM6A-LR',color='red')
    #plt.plot(x,N2OFut[1],color='cornflowerblue')
    
    plt.ylim(300,420)
    plt.axvline(x=2014,color='black',linestyle=':')
    plt.xlabel('Year') 
    if i == 1:
        plt.legend()
        plt.ylabel('N2O global mean mol mol-1')
    i=i+1

plt.savefig('N2O_ts_panel_10YearMean.eps')
plt.show()
        
```
