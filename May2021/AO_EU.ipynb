```python
Alldata=[]
AlldataEU=[]
models=['ERA5','EC-Earth3','IPSL-CM6A-LR','ACCESS-CM2','IITM-ESM',
            'MPI-ESM1-2-LR','MIROC6','MRI-ESM2-0','MPI-ESM1-2-HR',
        'CANESM5','INM-CM4-8','INM-CM5-0',]
for m in models :
    ts=xr.open_dataset('ERA5_telvars.nc')
    AO=ts.AO.values.tolist()
    EU=ts.EU.values.tolist()
    Alldata.append(AO)
    AlldataEU.append(EU)
    #print(AO)
```

```python

fig, ax = plt.subplots(figsize=(12, 6))
ax.violinplot(Alldata,
                  showmeans=True,
                  showmedians=False,
                  )
ax.set_xlabel('Models')
ax.set_ylabel('Observed Values')
ax.set_title('AO index')
ax.set_xticklabels(models,rotation=45.)
ax.yaxis.grid(True)
ax.set_xticks([y + 1 for y in range(len(Alldata))])
#ax.set_xticks( np.arange(len(models)))
#labels=models
#set_axis_style(ax, labels)
```

```python

fig, ax = plt.subplots(figsize=(12, 6))
ax.violinplot(AlldataEU,
                  showmeans=True,
                  showmedians=False,
                  )
ax.set_xlabel('Models')
ax.set_ylabel('Observed Values')
ax.set_title('EU index')
ax.set_xticklabels(models,rotation=45.)
ax.yaxis.grid(True)
ax.set_xticks([y + 1 for y in range(len(AlldataEU))])
#ax.set_xticks( np.arange(len(models)))
#labels=models
#set_axis_style(ax, labels)
```
